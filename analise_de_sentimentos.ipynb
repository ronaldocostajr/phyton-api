```python
import seaborn as sns

sns.histplot(data = df, x= 'word_count')
```

```python
def count_words(text_column):
  df['word_count'] = df[text_column].str.split().apply(len)

  min = df['word_count'].min()
  mean = df['word_count'].mean()
  max = df['word_count'].max()

  print('número minimo ', min )
  print('número médio ', mean)
  print('número maximo ', max)

  print('\nhistograma da frquencia de nueros de palavras')
  sns.histplot(data = df, x= 'word_count')
```

```python
count_words('content')
```

```python
score_count = df['score'].value_counts()
```

Verificando a disribuição dos scores

```python
sns.histplot(data = df, x= 'score', discrete = True)
```

```python
import matplotlib.pyplot as plt

plt.pie(score_count, labels = score_count.index, autopct='%1.1f%%')
plt.title('distribuição dos scores')
plt.axis('equal')
plt.show()
```

```python
def class_distribuition(class_column):
  class_counts = df[class_column].value_counts()

  print('número de instancia por classes')
  print(class_counts)

  print('\nhistograma')
  sns.histplot(data = df, x = class_column, discrete = True)
  plt.title('Histograma')
  plt.show()

  print('\npizza')
  plt.pie(class_counts, labels = class_counts.index, autopct='%1.1f%%')
  plt.title('distribuição dos scores')
  plt.axis('equal')
  plt.show()
```

```python
class_distribuition('score')
```

## Palavras mais frequentes

```python
from collections import Counter

all_text = ' '.join(df['content'])
words = all_text.split()
word_count = Counter(words)
most_comon_words = word_count.most_common(20)
most_comon_words_df = pd.DataFrame(most_comon_words, columns = ['palavra', 'frequencia'])
most_comon_words_df
```

```python
def most_frequency_words(df, text_column, number_words = 10):
  all_text = ' '.join(df[text_column])
  words = all_text.split()
  word_count = Counter(words)

  most_comon_words = word_count.most_common(number_words)
  most_comon_words_df = pd.DataFrame(most_comon_words, columns = ['palavra', 'frequencia'])
  display(most_comon_words_df)
```

```python
most_frequency_words(df, 'content', 50)
```
